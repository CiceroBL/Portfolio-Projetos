```python
import pandas as pd
```

```python
df = pd.read_csv("Video Games Sales.csv")
```

```python
df.head()
```

```python
df['Year'] = df['Year'].astype(str).replace('\.0', '', regex=True)
```

```python
df.head()
```

```python
df.dropna()
```

```python
df['Year'] = df['Year'].astype(str).replace('nan', '0', regex=True)
```

```python
df['Rest of World'] = df['Global'] - df['North America'] - df['Europe'] - df['Japan']
```

```python
df['North America'] = df['North America'].round(2)
df['Europe'] = df['Europe'].round(2)
df['Japan'] = df['Japan'].round(2)
df['Rest of World'] = df['Rest of World'].round(2)
df['Global'] = df['Global'].round(2)
```

```python
df.to_csv('games.csv', index=False)
```
